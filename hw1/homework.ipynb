## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
import os, json
import string
import re
import nltk
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
from nltk.corpus import stopwords

mystem = Mystem()
morph = MorphAnalyzer()
string.punctuation += '«»—…“”'
stops = stopwords.words('russian')
```

```python
f = open('text1.txt', "r", encoding="utf8")
pm = list()
ms = list()

text = f.read()
text_lst = list()

for word in text.split():
    text_lst.append(word.strip(string.punctuation))
text_lst = [word for word in text_lst if word not in stops]
```

```python
for token in text_lst:
    a=morph.parse(token)
    form=a[0].normal_form
    if form != ' ' and form != '\n' and  form != '':
        pm.append(form)
    else:
        continue
print(pm)
```

```python
mystem_rt = mystem.lemmatize(text)
for tkn in mystem_rt:
    tkn=tkn.strip(string.punctuation)
    if tkn != ' ' and tkn != '\n' and tkn != ' — ' and tkn != '':
        ms.append(tkn)
    else:
        continue

ms = [word for word in ms if word not in stops]
print(ms)
```

```python
print(len(ms),len(pm))

n = 0
for py in pm:
    if py.lower() in ms:
        n += 1
print(n)

m = 0
for my in ms:
    if my.lower() in pm:
        m += 1
print(m)
```

Различия:
- Mystem сохраняет результаты лемматизации в виде списка, в котором отдельными элементами выступают знаки препинания и пробелы; pymorphy показывает результаты в виде substrings, где можно видеть саму лемму и метаданные.
- Mystem сохраняет регистр элементов, а pymorphy приводит всё в нижний регистр.
- у pymorphy можно заметить некоторые ошибки (e.g. отлично!чтый).
- Mystem проводит лемматизацию значительно дольше, чем pymorphy
